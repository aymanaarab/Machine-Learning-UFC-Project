## 2. Analyse des valeurs aberrantes

Avant de choisir un facteur d'échelle, il est essentiel de comprendre l'ampleur des valeurs aberrantes.

Les caractéristiques présentant une forte asymétrie ou des valeurs extrêmes risquent de fausser les calculs de distance si elles ne sont pas prises en compte.

```python
# ── Skewness per feature ──────────────────────────────────────
skew = df[STYLE_FEATURES].skew().sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(10, 5))
colors = [ACCENT if abs(v) > 1.0 else GOLD if abs(v) > 0.5 else BLUE for v in skew.values]
bars = ax.barh(skew.index, skew.values, color=colors, alpha=0.85)
ax.axvline(0,   color='white', linewidth=0.8, alpha=0.5)
ax.axvline( 1,  color=ACCENT,  linewidth=1,   linestyle='--', alpha=0.6, label='|skew| > 1 (high)')
ax.axvline(-1,  color=ACCENT,  linewidth=1,   linestyle='--', alpha=0.6)
ax.axvline( 0.5,color=GOLD,    linewidth=1,   linestyle=':',  alpha=0.5, label='|skew| > 0.5 (moderate)')
ax.axvline(-0.5,color=GOLD,    linewidth=1,   linestyle=':',  alpha=0.5)
ax.set_title('Feature Skewness — Higher = More Outlier-Prone', fontsize=13)
ax.set_xlabel('Skewness')
ax.legend(fontsize=9)
plt.tight_layout()
plt.show()

high_skew = skew[skew.abs() > 1.0].index.tolist()
print(f'\nHighly skewed features (|skew| > 1): {high_skew}')
```

```python
# ── Boxplots for the most skewed features ─────────────────────
fig, axes = plt.subplots(2, 3, figsize=(16, 8))
fig.suptitle('Boxplots — Most Skewed Features (outliers visible)', fontsize=13)

top_skew = skew.abs().sort_values(ascending=False).head(6).index
palette  = [ACCENT, GOLD, BLUE, '#06d6a0', '#a786df', '#f4a261']

for ax, feat, color in zip(axes.flat, top_skew, palette):
    data = df[feat].dropna()
    ax.boxplot(data, vert=False, patch_artist=True,
               boxprops=dict(facecolor=color, alpha=0.6),
               medianprops=dict(color='white', linewidth=2),
               flierprops=dict(marker='o', markersize=2,
                               markerfacecolor=ACCENT, alpha=0.4))
    ax.set_title(feat, fontsize=10)
    ax.set_xlabel('Value')

plt.tight_layout()
plt.show()
```

# ***STANDARILISATION***



> **StandardScaler** utilise la moyenne et l'écart type.
Une seule valeur aberrante extrême (par exemple, un combattant avec 51 coups critiques/min) fausse la moyenne et augmente l'écart type, ce qui comprime les autres combattants dans une fourchette étroite.




> **RobustScaler** utilise la **médiane** et l'écart interquartile (EIQ).
Les valeurs aberrantes n'ont quasiment aucun effet sur la médiane ni sur l'EIQ ; ainsi, la plupart des combattants bénéficient d'une dispersion significative et les valeurs extrêmes ne dominent pas.




```python
# ── Visual proof: StandardScaler vs RobustScaler on sig_str_pm ─
feat = 'sig_str_pm'
raw  = df[feat].values.reshape(-1,1)

std_scaled    = StandardScaler().fit_transform(raw).flatten()
robust_scaled = RobustScaler().fit_transform(raw).flatten()

fig, axes = plt.subplots(1, 3, figsize=(16, 4))
fig.suptitle(f'Scaling Comparison on "{feat}"', fontsize=13)

for ax, data, title, color in zip(
    axes,
    [raw.flatten(), std_scaled, robust_scaled],
    ['Raw', 'StandardScaler', 'RobustScaler ✅'],
    [BLUE, GOLD, ACCENT]
):
    ax.hist(data, bins=50, color=color, alpha=0.85, edgecolor='none')
    ax.set_title(title, fontsize=11)
    ax.axvline(np.median(data), color='white', linestyle='--',
               label=f'median={np.median(data):.2f}')
    ax.legend(fontsize=8)

plt.tight_layout()
plt.show()

print('With RobustScaler: most fighters cluster near 0, outliers still present but contained')
```

#**#  Analyse en composantes principales (PCA)**

Nous avons 17 variables explicatives. L'PCA permet de réduire leur dimensionnalité tout en conservant la majeure partie de la variance.

Ceci permet :

- de réduire le bruit dû aux variables corrélées
- d'accélérer le clustering
- de permettre la visualisation 2D/3D

```python
pca_full = PCA()
pca_full.fit(X_scaled)

explained      = pca_full.explained_variance_ratio_
cumulative     = np.cumsum(explained)
n_components   = len(explained)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('PCA — Variance Explained', fontsize=13)

# Individual variance per component
ax1.bar(range(1, n_components+1), explained * 100,
        color=ACCENT, alpha=0.85, edgecolor='none')
ax1.set_title('Variance per Component')
ax1.set_xlabel('Principal Component')
ax1.set_ylabel('% Variance Explained')
ax1.yaxis.set_major_formatter(mticker.PercentFormatter())

# Cumulative variance
ax2.plot(range(1, n_components+1), cumulative * 100,
         color=GOLD, marker='o', linewidth=2, markersize=4)
ax2.fill_between(range(1, n_components+1), cumulative * 100, alpha=0.2, color=GOLD)
ax2.set_title('Cumulative Variance Explained')
ax2.set_xlabel('Number of Components')
ax2.set_ylabel('Cumulative % Variance')
ax2.yaxis.set_major_formatter(mticker.PercentFormatter())

# Mark key thresholds
for threshold, color, ls in [(0.80, BLUE, '--'), (0.90, '#06d6a0', ':')]:
    n = np.argmax(cumulative >= threshold) + 1
    ax2.axhline(threshold * 100, color=color, linestyle=ls, alpha=0.7,
                label=f'{int(threshold*100)}% → {n} components')
    ax2.axvline(n, color=color, linestyle=ls, alpha=0.7)

ax2.legend(fontsize=9)
plt.tight_layout()
plt.show()

print('\nVariance explained per component:')
for i, (ind, cum) in enumerate(zip(explained, cumulative)):
    marker = ' ← 80%' if abs(cum - 0.80) < 0.03 else (' ← 90%' if abs(cum - 0.90) < 0.03 else '')
    print(f'  PC{i+1:2d}: {ind*100:5.1f}%  |  cumulative: {cum*100:5.1f}%{marker}')
```

```python
# ── What does each component represent? ───────────────────────
# Look at the top contributing features per component
N_SHOW = 8   # components to inspect
loadings = pd.DataFrame(
    pca_full.components_[:N_SHOW].T,
    index=STYLE_FEATURES,
    columns=[f'PC{i+1}' for i in range(N_SHOW)]
)

fig, axes = plt.subplots(2, 4, figsize=(18, 9))
fig.suptitle('PCA Component Loadings — What Each Component Captures', fontsize=13)

for i, ax in enumerate(axes.flat):
    pc = f'PC{i+1}'
    vals = loadings[pc].sort_values()
    colors = [ACCENT if v < 0 else BLUE for v in vals.values]
    ax.barh(vals.index, vals.values, color=colors, alpha=0.85)
    ax.set_title(f'{pc}  ({explained[i]*100:.1f}% var)', fontsize=10)
    ax.axvline(0, color='white', linewidth=0.8, alpha=0.5)
    ax.tick_params(axis='y', labelsize=7)

plt.tight_layout()
plt.show()
```

```python
# ── Full scaled version ───────────────────────────────────────
X_full = X_scaled.copy()
print(f'X_full  shape : {X_full.shape}')

# ── PCA version (80% variance) ────────────────────────────────
N_PCA_COMPONENTS = np.argmax(cumulative >= 0.80) + 1
print(f'Components for 80% variance : {N_PCA_COMPONENTS}')

pca = PCA(n_components=N_PCA_COMPONENTS, random_state=42)
X_pca = pca.fit_transform(X_scaled)
print(f'X_pca   shape : {X_pca.shape}')
print(f'Variance retained : {pca.explained_variance_ratio_.sum()*100:.1f}%')

# Save PCA transformer
PCA_PATH = PROJECT_PATH +'/' +'pca_transformer.pkl'
with open(PCA_PATH, 'wb') as f:
    pickle.dump(pca, f)
print(f'\n✅ PCA transformer saved to {PCA_PATH}')
```

```python
# ── 2D PCA preview (for visualization only) ───────────────────
pca_2d = PCA(n_components=2, random_state=42)
X_2d   = pca_2d.fit_transform(X_scaled)

fig, ax = plt.subplots(figsize=(10, 7))
ax.scatter(X_2d[:, 0], X_2d[:, 1],
           alpha=0.4, s=15, color=ACCENT, edgecolors='none')
ax.set_xlabel(f'PC1 ({pca_2d.explained_variance_ratio_[0]*100:.1f}% variance)')
ax.set_ylabel(f'PC2 ({pca_2d.explained_variance_ratio_[1]*100:.1f}% variance)')
ax.set_title('2D PCA Projection — Fighter Style Space\n(color will be cluster labels in Notebook 4)',
             fontsize=12)
plt.tight_layout()
plt.show()

print(f'2D PCA captures {sum(pca_2d.explained_variance_ratio_)*100:.1f}% of total variance')
```
